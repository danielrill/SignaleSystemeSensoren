```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def conv(x):
    return x.replace(',', '.').encode()


ton = np.genfromtxt((conv(x) for x in open('.\\wilderton.csv')),
                     delimiter=';', skip_header=4, usecols=())

plt.plot(ton[:,0] , ton[:,1], 'b')
plt.ylabel('Spannung in mV')
plt.xlabel('Zeit in ms')
plt.grid(True)
plt.show()
```

```python
time = ton[:,0]

# AbtastIntervall
deltaT = np.abs(time[1] - time[0])
#print(deltaT)

fft = np.fft.fft(ton[:,1])

spektrum = np.abs(fft)

Signalänge = np.abs(np.max(ton[:, 0])) + np.abs(np.min(ton[:, 0]))
print("Signalänge " + str(Signalänge) + "\t in [ms]")

list= []
counter = 0
for i in np.array(ton[:,1]):
    
    list.append(i)

max = np.max(list)
min = np.min(list)

#print(max , min)

# Absolute Signaldauer
SignalDauer = np.abs(max) + np.abs(min)
np.round(SignalDauer, 1)
print("Signaldauer " + str(np.round(SignalDauer, 0) / 1000) + "\t in [s]")

#AbtastIntervall = SignalDauer/Signalänge
msek = 1000
AbtastIntervall = np.round((deltaT * msek), 2)
print("Abtastintervall " + str(AbtastIntervall) + "\t in [μs]")

Abtastfrequenz = 1 / deltaT
print("Abtastfrequenz " + str(np.round(Abtastfrequenz)) + "\t in [KHz]")


# n / (M * Δt)      Scwhingungen in freq umwandeln
freq = range(0, 10003, 1) / (Signalänge * deltaT)


plt.plot(freq, spektrum/1000)
plt.grid()
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude in mV')
plt.xlim(-20, 20000)
plt.show()

# For-Schleife um den passenden Frequenzwert zu erlangen der zu der maximalen Amplitude gehört
frequency = []
for x in range(0, 5000):
    if round(spektrum[x], 4) <= 7008.8274:
        frequency = freq[x]

print("Frequenz\t\t\t\t", np.round(np.mean(freq)))
print("Frequenz mit der größten Amplitude\t", np.round(np.max(spektrum))/1000 , "in [mV]")
print("Maximalster Amplitudenausschlag\t\t", np.round(np.max(frequency)/1000))
```

```python
# einlesen des kleinen Lautsprechers
klein = ["klein100HZ", "klein200HZ", "klein300HZ", "klein400HZ", "klein500Hz", "klein700HZ", "klein850HZ", "klein1KHZ",
         "klein1,2KHZ", "klein1,5KHZ", "klein1,7KHZ", "klein2KHZ", "klein3KHZ", "klein4KHZ", "klein5KHZ",
         "klein6KHZ", "klein10KHZ"]
# einlesen des großen Lautsprechers
gross = ["100HZ", "200HZ", "300HZ", "400HZ", "500Hz", "700HZ", "850HZ", "1KHZ",
         "1,2KHZ", "1,5KHZ", "1,7KHZ", "2KHZ", "3KHZ", "4KHZ", "5KHZ",
         "6KHZ", "10KHZ"]

# Vektoren für die Amplituden 
kleinamp = np.zeros(17)
grossamp = np.zeros(17)

kleinamp_avg = np.zeros(17)
grossamp_avg = np.zeros(17)

#--------------------------------
kleinamp_kov = np.zeros(17)
grossamp_kov = np.zeros(17)

kleinamp_std = np.zeros(17)
grossamp_std = np.zeros(17)

Bravais_Pearson_kl = np.zeros(17)
Bravais_Pearson_g = np.zeros(17)

#--------------------------------

for a in range(0, 17):
    # Einlesen der kleinen Dateien
    x = np.genfromtxt((conv(x) for x in open('.\\Messungen\\klein\\' + klein[a] + '.csv')),
                  delimiter=';', skip_header=4, usecols=())
    # Einlesen der großen Dateien
    u = np.genfromtxt((conv(x) for x in open('.\\Messungen\\' + gross[a] + '.csv')),
                  delimiter=';', skip_header=4, usecols=())

    # Berechnung für jede Datei des kleinen Lautsprechers den maximalen absoluten Amplitudenwert
    kleinamp[a] = np.max(np.abs(x[::,2]))
    kleinamp_avg[a] = np.mean(np.abs(x[::,2]))
    kleinamp_kov[a] = np.cov(np.abs(x[::,2]))
    #print(kleinamp[a], "test1")
    
    # Berechnung für jede Datei des großen Lautsprechers den maximalen absoluten Amplitudenwert
    grossamp[a] = np.max(np.abs(u[::,2]))
    grossamp_avg[a] = np.mean(np.abs(u[::,2]))
    grossamp_kov[a] = np.cov(np.abs(u[::,2])) 
    #print(grossamp[a]," test2")                  

    kleinamp_std[a] = np.std(np.abs(x[::,2]))
    grossamp_std[a] = np.std(np.abs(u[::,2]))

    Bravais_Pearson_kl[a] = kleinamp_kov[a]/(kleinamp_std[a] * grossamp_std[a] )
    Bravais_Pearson_g[a] = grossamp_kov[a]/(kleinamp_std[a] * grossamp_std[a] )

    # Die einzelnen .csv Dateien des kleinen Lautsprechers werden in jeweils einem Plot dargestellt
    plt.plot(x[::, 0], x[::, 2], 'r', linewidth=0.2)
    plt.title(klein[a])
    plt.ylabel('Spannung in V')
    plt.xlabel('Zeit t in s')
    plt.grid(True)
    #plt.savefig('data/img/' + str(klein[a]) + '.png')
    plt.show()

    # Die einzelnen .csv Dateien des großen Lautsprechers werden in jeweils einem Plot dargestellt
    plt.plot(u[::, 0], u[::, 2], 'g', linewidth=0.2)
    plt.title(gross[a])
    plt.ylabel('Spannung in V')
    plt.xlabel('Zeit t in s')
    plt.grid(True)
    #plt.savefig('data/img/' + str(gross[a]) + '.png')
    plt.show()
    # Die ersten beiden Zeiten für den kleinen Lautsprecher werden in time1 geschrieben
    time1 = x[::,0 ]
    # Die ersten beiden Zeiten für den großen Lautsprecher werden in time1 geschrieben
    time2 = u[::,0 ]

    # AbtastIntervall des kleinen Lautsprechers
    timing1 = round((time1[1] - time1[0]) * 100000, 5)

    # AbtastIntervall des großen Lautsprechers 
    timing2 = round((time2[1] - time2[0])  * 100000 , 5)


    # maximale Amplitudenwerte
    print("Amplitude\t\t", klein[a],"\t", np.max(np.abs(x[::,2]))) 
    print("AbtastIntervall\t\t\t\t", timing1, "ms")
    print("\n")
    print("Amplitude\t\t", gross[a],"\t\t", np.max(np.abs(u[::,2])))
    print("AbtastIntervall\t\t\t\t", timing2, "ms")

```
