```python
# Array um die einzelnen .csv Dateien einzulesen des kleinen Lautsprechers
klein = ["klein100HZ", "klein200HZ", "klein300HZ", "klein400HZ", "klein500Hz", "klein700HZ", "klein850HZ", "klein1KHZ",
         "klein1,2KHZ", "klein1,5KHZ", "klein1,7KHZ", "klein2KHZ", "klein3KHZ", "klein4KHZ", "klein5KHZ",
         "klein6KHZ", "klein10KHZ"]
# Array um die einzelnen .csv Dateien einzulesen des gro√üen Lautsprechers
gross = ["100HZ", "200HZ", "300HZ", "400HZ", "500Hz", "700HZ", "850HZ", "1KHZ",
         "1,2KHZ", "1,5KHZ", "1,7KHZ", "2KHZ", "3KHZ", "4KHZ", "5KHZ",
         "6KHZ", "10KHZ"]

# for-Schleife f√ºr die 17 unterschiedlichen .csv Dateien
for a in range(0, 17):
    # Einlesen der kleinen Dateien - dabei wird die linke Spalte als x und die rechte als y definiert
    x = np.genfromtxt((conv(x) for x in open('.\\Messungen\\klein\\' + klein[a] + '.csv')),
                  delimiter=';', skip_header=4, usecols=())
    # Einlesen der gro√üen Dateien - dabei wird die linke Spalte als u und die rechte als v definiert
    u = np.genfromtxt((conv(x) for x in open('.\\Messungen\\' + gross[a] + '.csv')),
                  delimiter=';', skip_header=4, usecols=())


    # Die einzelnen .csv Dateien des kleinen Lautsprechers werden in jeweils einem Plot dargestellt
    plt.plot(x[::, 0], x[::, 2], 'b', linewidth=0.2)
    plt.title(klein[a])
    plt.ylabel('Spannung in V')
    plt.xlabel('Zeit t in s')
    plt.grid(True)
    plt.figure(figsize=(17.5, 12))
    #plt.savefig('data/img/' + str(klein[a]) + '.png')
    plt.show()

    # Die einzelnen .csv Dateien des gro√üen Lautsprechers werden in jeweils einem Plot dargestellt
    plt.plot(u[::, 0], u[::, 2], 'g', linewidth=0.2)
    plt.title(gross[a])
    plt.ylabel('Spannung in V')
    plt.xlabel('Zeit t in s')
    plt.grid(True)
    plt.figure(figsize=(17.5, 12))
    #plt.savefig('data/img/' + str(gross[a]) + '.png')
    plt.show()
    # Die ersten beiden Zeiten f√ºr den kleinen Lautsprecher werden in time1 geschrieben
    time1 = x[::,0 ]
    # Die ersten beiden Zeiten f√ºr den gro√üen Lautsprecher werden in time1 geschrieben
    time2 = u[::,0 ]

    # Die zwei Zeiten des kleinen Lautsprechers werden subtrahiert und multipliziert
    # f√ºr eine mikrosekunden Darstellung
    # Zudem wird die berechnete Dauer gerundet
    timing1 = round((time1[1] - time1[0]) * 100000, 5)
    # Die zwei Zeiten des gro√üen Lautsprechers werden subtrahiert und multipliziert
    # f√ºr eine mikrosekunden Darstellung
    # Zudem wird die berechnete Dauer gerundet
    timing2 = round((time2[1] - time2[0]) * 100000, 5)

    # Ausgeben einer Tabelle im LaTeX Format
    print("/hline")
    # Der maximale Amplitudenwert des kleinen Lautsprechers f√ºr die jeweilige Frequenz
    print("Amplitude:", klein[a], np.max(np.abs(x[::,1])))
    # Der maximale Amplitudenwert des gro√üen Lautsprechers f√ºr die jeweilige Frequenz
    print("Amplitude:", gross[a], np.max(np.abs(u[::,1])))
    print("ms", klein[a], timing1)
    print("ms", gross[a], timing2)

print("/hline")
```

```python
def conv(x):
    return x.replace(',', '.').encode()


data = np.genfromtxt((conv(x) for x in open('.\\Messungen\\mitteltief.csv')),
                     delimiter=';', skip_header=4, usecols=())


time = data[::,0]


# Abtastintervall deltaT
difference = np.abs(time[1] - time[0])
print(difference)


data_fft = np.fft.fft(data[:, 1])

spektrum = np.abs(data_fft)

freq = range(0, len(data), 1) / (difference * len(data))

print(freq)

plt.plot(freq, spektrum/1000)
plt.grid()
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude in V')
plt.xlim(-10, 250)

plt.show()
```

```python
# Einlesen der Signall√§nge
file = open(".\\Messungen\\mitteltief.csv")

signallaenge = len(file.readlines())
# Sekundenumwandlung so wird 0,000001s zu 1¬µs
sek = 1000000
# Das Abtastintervall in ¬µs anzeigen und runden
abtastintervall = round((difference * sek), 2)

frequency = []
# For-Schleife um den passenden Frequenzwert zu erlangen der zu der maximalen Amplitude geh√∂rt
for x in range(0, 10000):                        # wenn wir 10003 nehmen -> 199.000
    if round(spektrum[x], 4) == 4892.2064:       # 7008.8274:         (Wilderton)
        frequency = freq[x]

# Berechnung der gr√∂√üten Amplitude
print("Grundperiode: 0.001275 s", )
print("Grundfrequenz: 784.31 Hz", )
print()
print("Signaldauer: ", (abtastintervall * signallaenge) / sek, " s")
print("Abtastfrequenz: ", 1 / abtastintervall * sek, " Hz")
print("Signall√§nge M: ", signallaenge)
print("Abtastintervall ‚àÜt:", abtastintervall, " Œºs")
print()
print("Maximalster Amplitudenausschlag", np.max(frequency) )
print("Frequenz mit der gr√∂√üten Amplitude", np.max(spektrum))
```
