```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

AUFGABE 1 

Bestimmung der Tonh√∂he eines akustischen Signals

```python
def conv(x):
    return x.replace(',', '.').encode()


wild = np.genfromtxt((conv(x) for x in open('.\\Messungen\\wilderton.csv')),
                     delimiter=';', skip_header=4, usecols=())

mittel = np.genfromtxt((conv(x) for x in open('.\\Messungen\\mitteltief.csv')),
                       delimiter=';', skip_header=4, usecols=())

fig, ax = plt.subplots()

ax.plot(wild[:, 0] * 1000, wild[:, 1] * 1000, color="green")
#ax.plot(mittel[:, 0] * 1000, mittel[:, 1] * 1000, color="orange")


fig.text(0.5, 0.04, 'Zeit in ms', ha='center', va='center')
fig.text(0.06, 0.5, 'Spannung in mw', ha='center',
         va='center', rotation='vertical')

plt.show()
```

```python
def conv(x):
    return x.replace(',', '.').encode()


data = np.genfromtxt((conv(x) for x in open('.\\Messungen\\wilderton.csv')),
                     delimiter=';', skip_header=4, usecols=())


data_fft = np.fft.fft(data[:,1])

freq = []

for i in range(len(data_fft)):
    freq.append( i / 0.01)

fig,ax = plt.subplots()

ax.plot(freq, np.abs(data_fft))

fig.text(0.5, 0.04, 'Frequenz (HZ)', ha='center', va='center')
fig.text(0.06, 0.5, 'Amplitude (V)', ha='center', va='center', rotation='vertical')

plt.show()

```

Aufgabe 2

Frequenzgang von Lautsprechern

```python
eingang = np.genfromtxt((conv(x) for x in open('.\\Messungen\\100HZ.csv')),
                   delimiter=';', skip_header=4, usecols=())

"""eingang = np.genfromtxt((conv(x) for x in open('.\\Messungen\\klein\\klein100HZ.csv')),
                        delimiter=';', skip_header=4, usecols=())"""
```

```python
for i in [1]:
    fig, ax = plt.subplots()

    ax.plot(eingang[:, 0], eingang[:, i])
    fig.text(0.5, 0.04, 'Frequenz (HZ)', ha='center', va='center')
    fig.text(0.06, 0.5, 'Amplitude (V)', ha='center',
             va='center', rotation='vertical')
             
    ax.grid(color='gray', linestyle=':', linewidth=0.5)
    plt.show()

for i in [2]:
    fig, ax = plt.subplots()

    ax.plot(eingang[:, 0], eingang[:, i] / 1000)
    fig.text(0.5, 0.04, 'Frequenz (HZ)', ha='center', va='center')
    fig.text(0.06, 0.5, 'Amplitude (V)', ha='center',
             va='center', rotation='vertical')
    
    ax.grid(color='gray', linestyle=':', linewidth=0.5)
    plt.show()
```

BODE-DIAGRAMM

```python
for i in [1]:
    fig, ax = plt.subplots()

    ax.plot(eingang[:, 0], 20 * np.log10(eingang[:, 1]))
    fig.text(0.5, 0.04, 'Frequenz (HZ)', ha='center', va='center')
    fig.text(0.06, 0.5, 'Amplitude in dB', ha='center',
             va='center', rotation='vertical')

    ax.grid(color='gray', linestyle=':', linewidth=0.5)

    plt.xlim(10**-1.5, 10**1.5)
    plt.ylim(-40, 1)

    plt.semilogx()
    plt.show()

for i in [2]:
    fig, ax = plt.subplots()

    ax.plot(eingang[:, 0], eingang[:, 2] / -1000000 * eingang[:,0] * 360)
    fig.text(0.5, 0.04, 'Frequenz (HZ)', ha='center', va='center')
    fig.text(0.06, 0.5, 'Phasenverschiebung in ms', ha='center',
             va='center', rotation='vertical')

    ax.grid(color='gray', linestyle=':', linewidth=0.5)

    plt.semilogx()
    plt.show()
```
